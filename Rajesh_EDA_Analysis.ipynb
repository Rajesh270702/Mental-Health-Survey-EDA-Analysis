# **Project Summary :**
This project involves a comprehensive Exploratory Data Analysis (EDA) of a mental health survey dataset, primarily focused on individuals working in the technology sector. The main objective of this analysis is to uncover patterns, trends, and potential risk factors influencing mental health in the workplace, with the ultimate goal of helping organizations make informed decisions about mental health support and policy design.

The dataset consists of 1,259 survey responses collected from professionals across various countries. It includes 27 features such as demographic details (age, gender, country, state), work-related attributes (company size, remote work status, availability of benefits), and personal experiences (treatment history, family history of mental illness, comfort in discussing mental health at work). The data was provided in a raw format (CSV), which required careful inspection and cleaning before analysis.

The EDA process was conducted using Python, with libraries like Pandas, Matplotlib, and Seaborn to handle data manipulation and visualization. The project began with a basic overview of the dataset, including checking the number of rows and columns, data types, and identifying missing values. The comments column, which had over 85% missing data, was dropped, while columns like self_employed were filled with a default value (‘No’). Duplicate entries were not present, but outlier detection was considered for age values.

Descriptive statistics revealed that the respondents were predominantly aged between 20 and 45 years, with a few extreme values (e.g., under 15 or over 100) likely being input errors. Gender distribution showed male dominance, but a notable number of respondents identified as female or other genders, highlighting the diversity in responses. Visualizations such as histograms, count plots, and heatmaps were used to understand the distribution and correlation between different features.

One of the key findings was a strong relationship between family history of mental illness and the likelihood of seeking treatment. Employees who had a family history of mental health conditions were significantly more likely to report receiving or needing treatment themselves. Another insight was the impact of company culture on mental health. Respondents who reported working in companies without mental health benefits or with poor leave policies showed more discomfort discussing mental health issues with coworkers or supervisors.

Remote work and company size also played important roles. While smaller companies were less likely to offer mental health support, they often provided a more personal work environment. Conversely, employees in larger companies reported more structured mental health policies but lower individual support.

The correlation heatmap indicated weak correlations among numerical variables, suggesting that mental health outcomes are more influenced by categorical and qualitative factors like support systems, awareness programs, and social environment.
